```python
import glob
import random
import matplotlib.pyplot as plt
import cv2
import os
import xml.etree.ElementTree as ET
from collections import defaultdict
```

```python
# paths
BASE_PATH = "../data/raw/NEU-DET"
SPLITS = ["train", "validation"]
```

```python
# function to extract defect class from XML
def parse_annotation(xml_file): 
    tree = ET.parse(xml_file)
    root = tree.getroot()
    labels = []
    for obj in root.findall("object"): 
        label = obj.find("name").text
        labels.append(label)
    return labels
```

```python
# default dictionary for counting defect types
label_counts = defaultdict(int)
```

```python
# fill the default dictionary
for split in SPLITS: 
    annotation_dir = os.path.join(BASE_PATH, split, "annotations")
    for xml_file in os.listdir(annotation_dir):
        if xml_file.endswith(".xml"):
            file_path = os.path.join(annotation_dir, xml_file)
            labels = parse_annotation(file_path)
            for label in labels: 
                label_counts[label] += 1
```

```python
# plot class distribution
plt.figure(figsize=(10, 5))
plt.bar(label_counts.keys(), label_counts.values(), color="mediumseagreen")
plt.title("Defect Type Distribution (train + validation)")
plt.xlabel("Defect Class")
plt.ylabel("Number of Samples")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# set up bounded boxes for the image display
def get_bounding_boxes(xml_path): 
    tree = ET.parse(xml_path)
    root = tree.getroot()
    bboxes = []
    labels = []
    for obj in root.findall("object"): 
        label = obj.find("name").text
        bbox = obj.find("bndbox")
        xmin = int(float(bbox.find("xmin").text))
        ymin = int(float(bbox.find("ymin").text))
        xmax = int(float(bbox.find("xmax").text))
        ymax = int(float(bbox.find("ymax").text))
        bboxes.append((xmin, ymin, xmax, ymax))
        labels.append(label)
    return bboxes, labels
```

```python
# load image paths recursively from subfolders
sample_dir = os.path.join(BASE_PATH, "train", "images")
image_paths = glob.glob(os.path.join(sample_dir, "**", "*.jpg"), recursive=True)

# sampe a few
sample_files = random.sample(image_paths, 6)
```

```python
# generate sample images for viewing
plt.figure(figsize=(15, 10))
for i, img_path in enumerate(sample_files): 
    # FIXED: use just the filename to find XML in flat annotations folder
    xml_filename = os.path.basename(img_path).replace(".jpg", ".xml")
    xml_path = os.path.join(BASE_PATH, "train", "annotations", xml_filename)

    if not os.path.exists(img_path) or not os.path.exists(xml_path): 
        print(f"❌ Missing file: {img_path} or {xml_path}")
        continue

    image = cv2.imread(img_path)
    if image is None: 
        print(f"❌ Failed to load image: {img_path}")
        continue

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    bboxes, labels = get_bounding_boxes(xml_path)
    for (xmin, ymin, xmax, ymax), label in zip(bboxes, labels): 
        cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
        cv2.putText(image, label, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

    plt.subplot(2, 3, i + 1)
    plt.imshow(image)
    plt.title(os.path.basename(img_path))
    plt.axis("off")

plt.tight_layout()
plt.show()
```
